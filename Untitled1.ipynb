### playing an audio

**librosa.display** 는 wave plot, spectrogram, colormap과 같은 다양한 포맷으로 오디오파일을 디스플레이 해준다

### spectogram

sound의 주파수들을 spectrum 형식으로 표현한 것

**.stft** 는 데이터를 short term fourier transform으로 converting

**.specshow** 는 display spectogram

```python
#display spectrogram
X=librosa.stft(x)
Xdb=librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb,sr=sr, x_axis='time', y_axis='hz')
#if to pring log of frequencies
librosa.display.specshow(Xdb, sr=sr, x_axis='time',y_axis='log')
plt.colorbar()
```

### Spectral centroid

**centre of mass**  weighted mean of the frequencies 

내용정리 필요



**.spectral_centroid** : frame마다의 spectral centroid값을 추출
    array return 값은 frame 수와 동일하게 개수의 column을 같는 배열을 return한다.
    
**.frames_to_time** frame->time으로 converting time[i]==frame[i]

```python
#spectral centroid --centre of mass --weighted mean of the frequencies present in the sound

import sklearn
spectral_centroids=librosa.feature.spectral_centroid(x,sr=sr)[0]
print(spectral_centroids.shape)
print(spectral_centroids)
#computing the time variable for visualization
frames=range(len(spectral_centroids))
t=librosa.frames_to_time(frames)


#normalising the spectral centroid for visualization

#normalize해주는 이유는 데이터를 더 쉽게 보기 위함 ->값을 0에서 1사이로 
def normalize(x,axis=0):
    return sklearn.preprocessing.minmax_scale(x,axis=axis)
#plotting the spectral centroid along the waveform
librosa.display.waveplot(x,sr=sr,alpha=0.4)
plt.plot(t, normalize(spectral_centroids),color='r')
```
